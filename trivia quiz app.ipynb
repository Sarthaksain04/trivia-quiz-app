```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk


class Question:
    def __init__(self, text, choices, correct_choice):
        self.text = text
        self.choices = choices
        self.correct_choice = correct_choice

class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Trivia Quiz")

        self.questions = []
        self.score = 0
        self.current_question_index = 0

        question1 = Question(
            "What is the capital of France?",
            ["London", "Paris", "Berlin", "Madrid"],
            "Paris"
        )

        question2 = Question(
            "Which planet is known as the 'Red Planet'?",
            ["Earth", "Mars", "Venus", "Jupiter"],
            "Mars"
        )
        
        question3 = Question(
            "Who played the  character of ironman in marvel movies ?",
            ["tom hillduston", "dwean jhonson", "chris hemsworth", "RDJ"],
            "RDJ"
        )
        question4 = Question(
            "what is the scientific name of orange ?",
            ["citrus raticulata", "pepever someniferum", "pisum sativum ", "dacus carota"],
            "citrus raticulata"
        )
        question5 = Question(
            " who is lavina ?",
            ["pagal ladki", "bina dimag ki ", "bhootni", "all options are correct"],
            "all option are correct "
        )
        self.questions.append(question1)
        self.questions.append(question2)
        self.questions.append(question3)
        self.questions.append(question4)
        self.questions.append(question5)

        self.question_label = tk.Label(root, text="", font=("Helvetica", 14))
        self.question_label.pack(pady=10)

        self.choice_buttons = []
        for i in range(4):
            button = tk.Button(root, text="", font=("Helvetica", 12), command=lambda i=i: self.check_answer(i))
            button.pack(fill=tk.X, padx=20, pady=5)
            self.choice_buttons.append(button)

        self.next_question_button = tk.Button(root, text="Next Question", font=("Helvetica", 12), command=self.next_question)
        self.next_question_button.pack(pady=10)

        self.update_question()

    def update_question(self):
        if self.current_question_index < len(self.questions):
            question = self.questions[self.current_question_index]
            self.question_label.config(text=question.text)
            for i, choice in enumerate(question.choices):
                self.choice_buttons[i].config(text=choice)
            self.next_question_button.config(state=tk.DISABLED)
        else:
            messagebox.showinfo("Quiz Complete", f"Quiz complete! Your score: {self.score}/{len(self.questions)}")
            self.root.quit()

    def check_answer(self, user_choice):
        question = self.questions[self.current_question_index]
        if question.choices[user_choice] == question.correct_choice:
            self.score += 1
        self.next_question_button.config(state=tk.NORMAL)

    def next_question(self):
        self.current_question_index += 1
        self.update_question()

def main():
    root = tk.Tk()
    app = QuizApp(root)
    root.title("trivia  Quiz App ")
    root["bg"]="cyan"
    root.geometry("600x400")
    root.mainloop()

if __name__ == "__main__":
    main()
```
